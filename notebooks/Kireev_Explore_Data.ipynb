## Дата публикации

```python
import json, datetime
```

```python
def read_date(f):
    line = json.loads(f)
    date = pd.Timestamp(line["published"]["$date"])
    return (line["_id"], date, date.dayofweek, date.hour)
```

```python
%%time
with open("../data/train.json") as f:
    lines = pd.DataFrame([read_date(line) for line in f.readlines()],
                         columns=["_id", "publ_date", "publ_dayofweek", "publ_hour"])
```

```python
date = lines.iloc[0].publ_date
date
```

# tags

```python
def read_tags(f):
    line = json.loads(f)
    tags = line["tags"]
    return (" ".join([tag.replace(" ", "_") for tag in tags]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
%%time
with open("../data/train.json") as f:
    X = cv.fit_transform([read_tags(line) for line in f.readlines()])
```
